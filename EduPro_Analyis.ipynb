```python
master=step1.merge(courses,on="CourseID",how="left")
master.head()
```

```python
master.shape[0]
master.shape[1]
master.columns.tolist()
```

```python
enrollments=master.groupby(["TeacherID","TeacherName","TeacherRating"]).size().reset_index(name="Enrollments")

top10=enrollments.sort_values("Enrollments",ascending=False).head(10)
top10
```

```python
correlation=teachers["YearsOfExperience"].corr(teachers["TeacherRating"])
correlation
```

```python
teachers["ExpBracket"]=pd.cut(
    teachers["YearsOfExperience"],
    bins=[0,3,7,12,25],
    labels=["0-3 yrs","4-7 yrs","8-12 yrs","13+ yrs"]
)

bracket_avg=teachers.groupby("ExpBracket",observed=True)["TeacherRating"].mean()
bracket_avg
```

```python
teachers["Tier"]=pd.cut(
    teachers["TeacherRating"],
    bins=[0,2,3.5,5],
    labels=["Low","Mid","High"]
)

enroll=master.groupby("TeacherID").size().reset_index(name="Enrollments")
teachers_enroll=teachers.merge(enroll,on="TeacherID",how="left")

tier_avg=teachers_enroll.groupby("Tier",observed=True)["Enrollments"].mean().reset_index()
tier_avg["Enrollments"]=tier_avg["Enrollments"].round(0)
tier_avg
```

```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
fig1=px.histogram(
    teachers,
    x="TeacherRating",
    nbins=20,
    title="Distribution of Teacher Ratings",
    labels={"TeacherRating": "Teacher Rating","count": "Number of Teachers"},
    color_discrete_sequence=["#2563EB"]
)

fig1.add_vline(
    x=teachers["TeacherRating"].mean(),
    line_dash="dash",
    line_color="red",
    annotation_text=f"Mean:{teachers['TeacherRating'].mean():.2f}"
)

fig1.update_layout(
        plot_bgcolor="white",
        title_font_size=18
)

fig1.show()
```

```python
fig2=px.bar(
    top10.sort_values("Enrollments"),
    x="Enrollments",
    y="TeacherName",
    orientation="h",
    title="Top 10 Teachers by Enrollments",
    labels={"TeacherName": "Teacher","Enrollments": "Total Enrollments"}, 
    color="TeacherRating", 
    color_continuous_scale="Blues", 
    text="Enrollments"
)

fig2.update_traces(textposition="outside")
fig2.update_layout(plot_bgcolor="white",title_font_size=18)

fig2.show()
```

```python
fig3=px.scatter(
    teachers,
    x="YearsOfExperience", 
    y="TeacherRating", 
    color="Gender", 
    size="TeacherRating", 
    hover_name="TeacherName", 
    trendline="ols", 
    title="Years of Experience vs Teacher Rating", 
    labels={"YearsOfExperience": "Years of Experience", 
           "TeacherRating": "Teacher Rating"}, 
    color_discrete_map={"Male": "#2563EB","Female": "#EC4899"}
)

fig3.update_layout(
    plot_bgcolor="white", 
    title_font_size=18
)

fig3.show()
```

```python
fig4=px.bar(
    category_rating.sort_values("Avg Rating"), 
    x="Avg Rating", 
    y="Category", 
    orientation="h", 
    title="Average Course Rating by Category", 
    color="Avg Rating", 
    color_continuous_scale="RdYlGn", 
    text="Avg Rating"
)

fig4.update_traces(textposition="outside")
fig4.update_layout(
    plot_bgcolor="white", 
    title_font_size=18, 
    coloraxis_showscale=False
)

fig4.show()
```

```python
fig5=px.bar(
    tier_avg, 
    x="Tier", 
    y="Enrollments", 
    title="Average Enrollments by Instructor Tier", 
    color="Tier", 
    color_discrete_map={
        "Low": "#EF4444", 
        "Mid": "#F59E0B", 
        "High": "#22C55E"
    }, 
    text="Enrollments"
)
fig5.update_traces(textposition="outside")
fig5.update_layout(
    plot_bgcolor="white" , 
    title_font_size=18, 
    showlegend=False
)

fig5.show()
```

```python
heatmap_data=courses.groupby(
    ["CourseCategory","CourseLevel"]
)["CourseRating"].mean().unstack().round(2)

fig6=px.imshow(
    heatmap_data, 
    color_continuous_scale="RdYlGn", 
    zmin=1,zmax=5, 
    title="Course Rating Heatmap: Category vs Level", 
    labels={"x":"Course Level","y": "Category","color": "Avg Rating"}, 
    text_auto=True, 
    aspect="auto"
)

fig6.update_layout(
    title_font_size=18, 
    width=700, 
    height=500, 
    xaxis=dict(side="bottom")
)

fig6.show()
```
